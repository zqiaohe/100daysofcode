Задача 2: аппроксимация функции
Рассмотрим сложную математическую функцию на отрезке [1, 15]:

```
def f(x):
  return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```
x = np.arange(0, 16, 0.1)
y = f(x)
```

```
plt.figure()
plt.plot(x,y)
plt.show()
```

```
plt.figure(figsize=(13,8))
plt.plot(x,y, 'r')
plt.xlabel('Ось иксов или абсцисс')
plt.ylabel('Ось игеков или ординат')
plt.title('Суперграфик из домахи')
plt.show()
```

```
import numpy
a11, a12, b1 = [float(i) for i in input().split()]
a21, a22, b2 = [float(i) for i in input().split()]
M1 = numpy.array([[a11, a12], [a21, a22]]) # Матрица (левая часть системы)
v1 = numpy.array([b1,b2]) # Вектор (правая часть системы)
r = numpy.linalg.solve(M1,v1) #Находим решение системы
print (r[0], r[1])
```

```
import numpy
a11, a12, b1 = [float(i) for i in input().split()]
a21, a22, b2 = [float(i) for i in input().split()]
M1 = numpy.array([[a11, a12], [a21, a22]]) # Матрица (левая часть системы)
v1 = numpy.array([b1,b2])
if numpy.linalg.det(M1)==0:
    print("Система не имеет решений")
else:
    r = numpy.linalg.solve(M1,v1) #Находим решение системы
    print (r[0], r[1])
```

```
import numpy
a11, a12, a13, b1 = [float(i) for i in input().split()]
a21, a22, a23, b2 = [float(i) for i in input().split()]
a31, a32, a33, b3 = [float(i) for i in input().split()]
M1 = numpy.array([[a11,a12,a13],[a21,a22,a23],[a31,a32,a33]])
v1 = numpy.array([b1,b2,b3])
if numpy.linalg.det(M1)==0:
    print("Система не имеет решений")
else:
    r = numpy.linalg.solve(M1,v1) #Находим решение системы
    print(r[0], r[1], r[2])
```
